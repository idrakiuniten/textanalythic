```python
#NAME:NUR IDRAKI HAFIY BIN MOHAMAD ZAIDI 
#ID:IS01084550
#NAME:CHARANPAL SINGH A/L MANJEET SINGH
#ID:IS01083868

# Lab Assignment 1
# Platform: quotes.toscrape.com


# Import packages
import requests
import pandas as pd
from bs4 import BeautifulSoup
from datetime import datetime
import time


# Base URL 

base_url = "http://quotes.toscrape.com/page/"

# Number of pages (Limit to 5 as required)
num_pages = 5

# List to store scraped data
all_reviews = []


# Function: scrape_page

def scrape_page(page_number):
    
    url = base_url + str(page_number) + "/"
    
    try:
        response = requests.get(url, timeout=10)
        
        if response.status_code != 200:
            print(f"Failed to retrieve page {page_number}")
            return
        
        soup = BeautifulSoup(response.text, "html.parser")
        
        # Find all quote blocks (treat as reviews)
        reviews = soup.find_all("div", class_="quote")
        
        for review in reviews:
            
            # Extract Reviewer Name (Author)
            reviewer_name = review.find("small", class_="author").text
            
            # Extract Review Content (Quote text)
            review_content = review.find("span", class_="text").text
            
            # Generate Review Date
            review_date = datetime.now().strftime("%Y-%m-%d")
            
            # Store data
            all_reviews.append({
                "Reviewer Name": reviewer_name,
                "Review Date": review_date,
                "Review Content": review_content
            })
        
        print(f"Page {page_number} scraped successfully")
        time.sleep(1)
        
    except Exception as e:
        print(f"Error on page {page_number}:", e)

# ------------------------------------------
# Scrape 5 Pages
# ------------------------------------------
for page in range(1, num_pages + 1):
    scrape_page(page)

# ------------------------------------------
# Save to CSV
# ------------------------------------------
df = pd.DataFrame(all_reviews)

file_name = f"Lab1Assignment_reviews_{datetime.now().strftime('%Y%m%d_%H%M%S')}.csv"
df.to_csv(file_name, index=False)

print("\nScraping Completed")
print("Total Reviews:", len(df))
print("Saved as:", file_name)
```
